# Text Generation

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/amazon_data/description_keyphrases_v1.csv',index_col=0)
```

## 1) fine tuning 한 모델로 text 생성

```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='/content/drive/MyDrive/amazon_data/model2/checkpoint-1000')
set_seed(42)
result = list()
for i in range(len(data)):
  result.append(generator(data["nlp_keywords"][i], max_length=30, num_return_sequences=1)[0]["generated_text"])
result
```

```python
data.loc[:,'description'] = result
```

```python
data.to_csv("/content/drive/MyDrive/amazon_data/result.csv")
```

## 2) pretrained GPT-2 모델 사용하여 text 생성(finetuning 안한 것)

```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model="gpt2")
set_seed(42)
result2 = list()
for i in range(len(data)):
  result2.append(generator(data["nlp_keywords"][i], max_length=30, num_return_sequences=1)[0]["generated_text"])
result2
```

```python
data.loc[:,'description'] = result2
```

```python
data.to_csv("/content/drive/MyDrive/amazon_data/result2.csv")
```
